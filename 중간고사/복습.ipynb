```python
import cv2

url = 'https://192.168.0.2:8080/video'
cap = cv2.VideoCapture(url)

frame_size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),
              int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
print(frame_size)

while True:
    ret, frame = cap.read()
    
    frame = cv2.resize(frame, (720, 480))
    
    if not ret:
        break
    
    cv2.imshow('frame', frame)
    
    key = cv2.waitKey(25)
    if key == 27:
        break
    
if cap.isOpened():
    cap.release()
cv2.destroyAllWindows()
```

### 비디오 녹화

```python
import cv2

cap = cv2.VideoCapture(0)
frame_size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),
              int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
print(frame_size)

fps = 20.0

fourccd = cv2.VideoWriter_fourcc(*'DIVX')
fourccp = cv2.VideoWriter_fourcc(*'PIM1')
fourccmj = cv2.VideoWriter_fourcc(*'MJPG')
fourccx = cv2.VideoWriter_fourcc(*'XVID')
fourccmp = cv2.VideoWriter_fourcc(*'MPEG')
fourccx = cv2.VideoWriter_fourcc(*'X264')

out1 = cv2.VideoWriter('./r1.mp4', fourccd, fps, frame_size)
out2 = cv2.VideoWriter('./r2.mp4', fourccp, fps, frame_size)
out3 = cv2.VideoWriter('./r3.mp4', fourccmj, fps, frame_size)
out4 = cv2.VideoWriter('./r4.mp4', fourccx, fps, frame_size)
out5 = cv2.VideoWriter('./r5.mp4', fourccmp, fps, frame_size)
out6 = cv2.VideoWriter('./r6.mp4', fourccx, fps, frame_size)

record = False

while True:
    ret, frame = cap.read()
    
    if not ret:
        break
    
    cv2.imshow('frame', frame)
    
    key = cv2.waitKey(25)
    
    if key == 27:
        break
    
    if key == ord('v'):
        record = True
        
    if key == ord('b'):
        out1.release()
        out2.release()
        out3.release()
        out4.release()
        out5.release()
        out6.release()
        record = False
    
    if record:
        out1.write(frame)
        out2.write(frame)
        out3.write(frame)
        out4.write(frame)
        out5.write(frame)
        out6.write(frame)

cv2.destroyAllWindows()
```

### 마우스 ROI

```python
import cv2

src = cv2.imread('c:/data/lena.jpg', cv2.IMREAD_GRAYSCALE)
roi = cv2.selectROI(src)
print(roi)

if roi != (0, 0, 0, 0):
    img = src[roi[1]:roi[1] + roi[3],
              roi[0]:roi[0] + roi[2]]
    
    cv2.imshow('img', img)
    cv2.waitKey()
    
cv2.destroyAllWindows()
```

### 임계값 threshold

```python
import cv2
from matplotlib import pyplot as plt

src = cv2.imread('c:/data/lena.jpg', cv2.IMREAD_GRAYSCALE)

tresh = 100
_, src2 = cv2.threshold(src, tresh, 255, cv2.THRESH_BINARY)
_, src3 = cv2.threshold(src, tresh, 255, cv2.THRESH_TOZERO)

plt.imshow(src2, cmap='gray')
plt.show()

plt.imshow(src3, cmap='gray')
plt.show()
```

### 적응형 임계값

```python
import cv2

src = cv2.imread('c:/data/lena.jpg', cv2.IMREAD_GRAYSCALE)

ret, dst = cv2.threshold(src, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

dst2 = cv2.adaptiveThreshold(src, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 51, 7)

dst3 = cv2.adaptiveThreshold(src, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 51, 7)

plt.imshow(src, cmap='gray')
plt.show()

plt.imshow(dst, cmap='gray')
plt.show()

plt.imshow(dst2, cmap='gray')
plt.show()

plt.imshow(dst3, cmap='gray')
plt.show()
```

### 히스토그램 평활화

```python
import cv2
from matplotlib import pyplot as plt

src = cv2.imread('c:/data/lena.jpg')

src2 = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
gray = cv2.equalizeHist(src2)

src3 = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(src3)
v2 = cv2.equalizeHist(v)
hsv = cv2.merge([h, s, v2])
src3 = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)

plt.imshow(gray, cmap='gray')
plt.show()

plt.imshow(src3)
plt.show()
```
